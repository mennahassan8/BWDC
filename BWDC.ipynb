```python
import pandas as pd
```

```python
edu=pd.read_csv('edu_01.csv')
```

```python
edu
```

```python
edu['race_or_ethnicity'].value_counts()
```

```python
edu.groupby(['race_or_ethnicity'])['rate'].mean()
```

```python
edu.groupby(['race_or_ethnicity', 'year'])['rate'].mean()
```

```python
edu.info()
```

```python
asian_edu=edu[(edu['race_or_ethnicity'] == 'Asian')][['race_or_ethnicity', 'year', 'rate']]
asian_edu
```

```python
asian_edu_grouped=asian_edu.groupby(['race_or_ethnicity', 'year'])['rate'].mean().reset_index()
asian_edu_grouped
```

```python
asian_edu_grouped_plot=asian_edu_grouped.plot(kind='bar', x='year', y='rate', color=['#3ed213','#10b410'], legend=False)
asian_edu_grouped_plot.set_title('Asian Degree Attainment Rate')
asian_edu_grouped_plot.set_xlabel('Year')
asian_edu_grouped_plot.set_ylabel('Rate');
```
